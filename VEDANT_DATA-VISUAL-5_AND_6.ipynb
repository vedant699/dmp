```python
import matplotlib.pyplot as plt
#BASIC LINE GRAPH USING PLOT()
plt.plot([1,2,3],[2,4,2])
plt.show()
import numpy as np
ypoints=np.array([2,4,1,5])
 #plotting without x-points
plt.figure()
plt.plot(ypoints)
plt.show()
#color changing to red
plt.plot(ypoints,'g')
plt.show()
plt.plot(ypoints,'o')
plt.show()
#marker
plt.plot(ypoints,marker='o',ms=15,mec='r',mfc='y')
plt.show()
plt.plot(ypoints,'o:r')
```

```python
import matplotlib.pyplot as plt
import numpy as np
#multiple lines
x=[10,4,7,1]
y=[2,8,3,9]
plt.plot(x,marker='^')
plt.plot(y,marker='D')
plt.show()
plt.grid(axis='x')
plt.grid(axis='y')
```
